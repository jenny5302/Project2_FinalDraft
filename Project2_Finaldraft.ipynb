### Scientific Question: What are the sequential and structral differences between human Tau protein and its evolutionary closest relative? 

The tau proteins are a group of six highly soluble protein isoforms produced by alternative splicing from the gene MAPT (microtubule-associated protein tau). They have roles primarily in stabilizing the internal microtubules in axons and are abundant in the neurons of the central nervous system (CNS). Tau dysfunction has been identified in multiple different brain diseases like Alzheimer's and Parkinson's Disease. The modified tau proteins form neurofibrillary tangles in the brain of AD patients. (Ellison, 2021) In Alzheimer's disease, the neuronal cytoskeleton in the brain is progressively disrupted and replaced by tangles of paired helical filaments (PHF) and straight filaments, mainly composed of hyperphosphorylated forms of TAU (PHF-TAU or AD P-TAU).

The two primary cardinal lesions associated with Alzheimer's disease are the neurofibrillary tangle and the amyloid plaque. The neurofibrillary tangle consists of abnormal accumulations of abnormally phosphorylated Tau protein. The amyloid plaque consists of a central core of beta-amyloid, surrounded by abnormally configured neuronal processes or neurites. (Perl, 2010) Squirrel monkeys, vervets, lemurs, apes, other nonhuman primates, and even dogs develop copious amyloid plaques as they age. However, most of the animals are spared of the neurofibrillary tangles formed by Tau. The previous study has shown a co-presence of PHFs and Aβ-amyloidosis in Chimpanzees. (Rosen et al., 2008) Another study has found abnormally phosphorylated tau protein in sheep and goat brains. (Braak et al., 1994) Based on these studies, I'm particularly interested in the relationship between the susceptibility to Alzheimer's disease, especially neurofibrillary tangle formation, and the structure of Tau protein.

### Part 1A: Multiple Sequence alignment

A FASTA file is a text file. Each sequence begins with a single-line description, followed by lines of sequence data. The single-line description contains a greater-than (>) symbol in the first column, followed by the sequence name. 

Multiple sequence alignment(MSA) refers to the sequence alignment of three or more biological sequences, in this case protein sequence. Here, MSA is performed through clustal omega, using the fasta file downloaded from Uniprot which contains the 11 unaligned sequences. The aligned sequences of the 11 tau proteins are stored in a fasta file.<br>

```python
# Write the multiple sequence alignment as a function

# Create a function called COalign
# Takes in three parameters
# filein is the file name containing the sequences to be aligned
# fileout is the file name and type where the function will store the output(aligned sequences)
# gtout is the file name and type where the function will store the guide tree output file
# the guide tree file will later be used to generate phylogenetic tree
def COalign(filein, fileout, gtout):
    
    # Use Clustal Omega Command line function to perform multiple sequence alignment
    # Define local variable CO_cline, which later becomes a function that can be called
    CO_cline = ClustalOmegaCommandline(infile = filein, # passing the parameter filein into the function
                                       
                                       # passing fileout into the function
                                      outfile = fileout, 
                                       
                                       # this line will force file overwrite
                                       #pretending errors from running the function multiple times
                                      force = True, 
                                       
                                       # this line tells clustal omega to use 
                                       # full distance matrix for guide-tree calculation
                                      distmat_full = True,
                                       
                                       # passing gtout into the function
                                      guidetree_out = gtout,
                                       
                                       # tell clustal omega to store the output as fasta file format
                                      outfmt = 'fasta')
    
    # perform MSA based on the settings we passed into the function 
    CO_cline()
    
    # visualize the alignment by using the read function in AlignIO package
    # Define local variable align, which contains the result of MSA
    align = AlignIO.read(fileout, 'fasta')
    
    # Return the printout sequences
    return print(align)
```

```python
# Call the COalign function to perform MSA and visualize the aligned sequences
COalign('Data/Tau_unaligned.fasta', 'Data/Tau_aligned.fasta', 'Data/Tau_aligned.dnd')
```

### Part 2A: Phylogenetic tree (Plotting the result of multiple sequence alignment)

Phylogenetic trees can be used to represent hypothesized evolutionary relationships among organisms by using nodes to represent a most recent common ancestor between species andusing a bar to represent the “amount of genetic change” for a given branch length relative tothe given phylogenetic tree.<br>
In this case, the MSA of the 10 tau protein sequences are used to draw a phylogenetic tree to represent the evolutionary relationship between these proteins. I take the MSA result (newick tree file) from above and draw a phylogenetic tree based on it.

```python
# Use Phylo.read to generate Phylogenetic tree
# Read in the guide tree output file from MSA, the file type is newick tree file
# Define global variable tree, containing the parsed guide tree file
tree = Phylo.read("Data/Tau_aligned.dnd", "newick")

#The function draw_ascii creates a simple ASCII-art (plain text) dendrogram.
# It takes in the parsed guide tree file and generate a phylogenetic tree based on that
Phylo.draw_ascii(tree)
```

### Part 1B & 2B: Homology modeling

Homology modeling, also known as comparative modeling of protein, refers to constructing an atomic-resolution model of the "target" protein from its amino acid sequence and an experimental three-dimensional structure of a related homologous protein (the "template"). (Wikipedia)

In my project, I will use SWISS-model, which is a fully automated protein structure homology-modeling server. (https://swissmodel.expasy.org/) On this website, I will use the User-template input type (https://swissmodel.expasy.org/interactive#structure), which takes in a target sequence in either plain string or fasta file type, and a template file, which is a pdb file downloaded from the Protein Data Bank. Then the website will generate a homology model for my target sequence, along with the Qmean Z score plot, and local quality estimate (Explanation in following sections).

I found a NMR structure of human Tau protein, showing the structure of it when it is bound to microtubules. I used this structure as a template, and the TAU_PANTR sequence from UniProt as the target sequence to generate a homology model of the normal chimpanzee tau protein. 

The diseased template that I chose is a cryo-EM structure of the paired helical filament from AD brain. I used this as a template and the TAU_PANTR sequence as the target sequence.

#### Explanation of the output from homology modeling
- Local Quality Estimate shows the expected similarity to the native structure (y axis) for each residue of the model (x axis). The bars above the line indicate relaticely good estimate for the target structure while the bars below the line indicate low quality estimate. The blue color indicates high confidence gradient and the orange color represents low confidence gradient.
- GMQE gives an overall model quality measurement between 0 and 1, with higher numbers indicating higher expected quality.
- The QMEAN Z score shows how many SD from the mean is my model score given a score distribution from a large set of experimentally determined structures. Z-scores around 0.0 therefore reflect a "native-like" structure.
- The QMEAN Z score of the model in comparison with non-redundant set of PDB Structures is shown in a dot plot. The x-axis shows protein length (number of residues). The y-axis is the "QMEAN" score. Every dot represents one experimental protein structure. Black dots are experimental structures with a "QMEAN" score within 1 standard devation of the mean (|Z-score| between 0 and 1), experimental structures with a |Z-score| between 1 and 2 are grey. Experimental structure that are even further from the mean are light grey. The target model is represented as a red star.

This is the result of PANTR_healthy homology modeling. 

The sequence identity between the target and the template is 100%. The Global Model Quality Estimate(GMQE) score for the target model is 0.65. The QMEAN Z score is -4.96. These scores show that PANTR_healthy model is of low quality and might not be able to represent the nature structure accurately.

Here we can see from the Local Quality Estimate that the overall quality of the estimate is not good for the chimpanzee tau protein. Most of the residues predicted are in the orange color with a mean of 0.53. Generally, this score indicates that the homology model is of low quality. 

From the dot plot, we can see that there are some gray dots near the PANTR_healthy model, showing that it can still be within 1-2 SD from the mean. However, it is indeed far away from the mean, falling in a region with few dots, meaning that it might not be reliable enough.
